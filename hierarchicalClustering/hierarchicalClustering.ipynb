```python
# Libraries
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.metrics import silhouette_score
plt.style.use("seaborn")
```

The two chunks below create figures for the article.

```python
# Create synthetic clusters
X, y = make_blobs(n_samples=10, centers=3, n_features=2, random_state=123)

# Figure 1 -- shows dendrogram
Z = linkage(X, 'complete')
plt.figure(figsize=(10, 6))
plt.title('Dendrogram', fontdict = {'fontsize' : 16})
plt.xlabel('Observation', fontdict = {'fontsize' : 14})
plt.ylabel('Distance', fontdict = {'fontsize' : 14})
dendrogram(
    Z,
    leaf_font_size=16.,  # font size for the x axis labels
)
#plt.show()
plt.savefig('figures/dendrogram.eps', format='eps')
```

```python
# Figure 2 -- shows cluster assignment based on cut
Z = linkage(X, 'complete')
plt.figure(figsize=(10, 6))
plt.title('Dendrogram with Horizontal Cuts', fontdict = {'fontsize' : 16})
plt.xlabel('Observation', fontdict = {'fontsize' : 14})
plt.ylabel('Distance', fontdict = {'fontsize' : 14})
plt.axhline(y=8, c='k', linestyle='--')
plt.axhline(y=5, c='k')
dendrogram(
    Z,
    leaf_font_size=16.,  # font size for the x axis labels
)
#plt.show()
plt.savefig('figures/dendrogram_cuts.eps', format='eps')
```

Now we move on to the implementation in Python example. 

```python
# Create synthetic clusters
X, y = make_blobs(n_samples=50, centers=2, n_features=2, cluster_std=2, random_state=123)

# Plot
plt.scatter(X[:,0], X[:,1], c=y, cmap='rainbow')
plt.title("True cluster labels")
plt.xlabel("x0")
plt.ylabel("x1")
plt.show()

#X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123, test_size=0.25)
```

```python
# Assign linkage
Z = linkage(X, 'complete')

plt.figure(figsize=(10, 8))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
fcluster(Z, t=3, criterion='maxclust')
```

```python
#silhouette_score(X, fcluster(Z, t=2, criterion='maxclust'))
def sil_scores(X, Z, ts):
    scores = []
    
    for num_clust in ts:
        scores.append(silhouette_score(X, fcluster(Z, t=num_clust, criterion='maxclust')))
    
    return scores

ts = [2,3,4,5]
scores = sil_scores(X, Z, ts)

for t, s in zip(ts, scores):
    print(f"With {t} clusters, the silhouette score is {s}")
```

```python
fcluster(Z, t=3, criterion='maxclust')
```

```python
# Plot
plt.scatter(X[:,0], X[:,1], c=fcluster(Z, t=3, criterion='maxclust'), cmap='rainbow')
plt.title("True cluster labels")
plt.xlabel("x0")
plt.ylabel("x1")
plt.show()
```
